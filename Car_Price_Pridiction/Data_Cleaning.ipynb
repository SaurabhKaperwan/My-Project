```python
import pandas as pd
import numpy as np
```

```python
car_raw = pd.read_csv('Car_Raw_Data.csv')
car_raw.head(10)
```

```python
car_raw.describe(include='all')
```

```python
car_raw.drop('Model',axis=1,inplace=True)
```

```python
car_raw.isnull().sum()
```

```python
car_raw.dropna(axis=0,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,4))
sns.displot(car_raw['Price'])
plt.title('Car Price Distribution')
```

```python
q=car_raw['Price'].quantile(q=0.99)
q
```

```python
car_1=car_raw[car_raw['Price']<q]
car_1.describe()
```

```python
plt.figure(figsize=(10,10))
sns.displot(car_1['Mileage'])
```

```python
q=car_1['Mileage'].quantile(0.99)
car_2=car_1[car_1['Mileage']<q]
```

```python
plt.figure(figsize=(100,1))
sns.displot(car_1['EngineV'])
```

```python
car_3=car_2[car_2['EngineV']<6.5]
```

```python
plt.figure(figsize=(8,4))
sns.displot(car_1['Year'])
```

```python
q=car_3['Year'].quantile(0.01)
car_4=car_3[car_3['Year']>q]
```

```python
car_4.describe(include='all')
```

```python
car_4
```

```python
car_clean=car_4.reset_index(drop=True)
```

```python
car_clean
```

```python
car_clean.to_csv('Car_cleaned.csv',index=False)
```
